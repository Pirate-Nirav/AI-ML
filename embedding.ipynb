```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ['LANGCHAI_API_KEY'] = os.getenv("LANGCHAI_API_KEY")
```

```python
from langchain_google_genai import  GoogleGenerativeAIEmbeddings
embedding = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-exp-03-07")
embedding
```

```python
text = " I am using google embeddings"
text_result = embedding.embed_query(text)
text_result
```

```python
len(text_result)
```

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader('data\\speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
## Vector Embedding And Vector StoreDB
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents,embedding)
db
```

```python
### Retrieve the results from query vectorstore db
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```

```python
for i, doc in enumerate(retrieved_results, 1):
    print(f"\nResult {i}:")
    print(f"Source: {doc.metadata.get('source')}")
    print(f"Content: {doc.page_content}")
```
