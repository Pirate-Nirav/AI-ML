Removing Stop Words

Stop words” are the foremost common words during a language like “the”, “a”, “on”, “is”, “all”. These words don't carry important meaning and are usually faraway from texts. It is possible to get rid of stop words using tongue Toolkit (NLTK), a set of libraries and programs for symbolic and statistical tongue processing.

### **Stemming**
Stemming is the process of reducing a word to it's word stem that affixes and prefixes or to the roots of the words known as a lemma. Stemming is important in nlu(Natural Language Understanding) and nlp.

### **RegexpStemmmer class**

NLTK has regexpstemmer class with the help of which we can easily implement regular expression stemmer algorithms. It basically takes a single regular expression and removes prefix and sufix that matches the expression.

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min = 4)
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

```python
message.rename(columns={"v1":"label","v2":"message"}, inplace=True)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))
```

```python
corpus = []
for i in range(0,len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```
